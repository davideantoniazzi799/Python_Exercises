DATE E ORE NEI GRAFICI:
- formato data datetime(anno, mese, giorno), creare grafico base
- aggiungere linestyle, rendere date più leggibili, cambiare formato date

```python
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
from datetime import datetime, timedelta

#ci serve importare le seguenti librerie

plt.style.use("ggplot")
```

CREARE GRAFICO BASE

```python
#il comando è plt.plot_date(x,y) dove le x sono le nostre date

x = [
    datetime(2022,1,10),
    datetime(2022,1,15),
    datetime(2022,1,20),
    datetime(2022,1,25),
    datetime(2022,1,30)
] #questo sono le nostre date

y = [3,6,2,8,5]

plt.plot_date(x,y)

plt.title("Nome grafico")
plt.show
```

PERSONALIZZAZIONE

```python
x = [
    datetime(2022,1,10),
    datetime(2022,1,15),
    datetime(2022,1,20),
    datetime(2022,1,25),
    datetime(2022,1,30)
] #questo sono le nostre date

y = [3,6,2,8,5]

#con linestyle = "tipo_linea" possiamo aggiungere una linea

#gcf sta per Get_Current_Figure e ci serve per scrivere le date in un formato migliore

#gca sta per Get_Current_Axis. Dobbiamo indicare l'asse che vogliamo(x in questo caso)
#poi mettiamo .set_major_formatter() dove inseriamo la data formattata
#le date formattate vanno create in precedenza

plt.plot_date(x,y, linestyle ="solid")
plt.gcf().autofmt_xdate()

date_formattate = mpl_dates.DateFormatter("%d %b %Y") #d=day, b=mese scritto, Y=anno
plt.gca().xaxis.set_major_formatter(date_formattate)

plt.title("Nome grafico")
plt.show
```
