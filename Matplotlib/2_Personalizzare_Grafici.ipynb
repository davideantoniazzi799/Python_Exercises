PERSONALIZZARE GRAFICI:
- titolo, etichette, seconda linea, legenda
- griglia, colore, dimensione, stile linea, marcatori, stile grafico

```python
from matplotlib import pyplot as plt
```

```python
x = [25, 26, 27, 28, 29, 30]
y1 = [100, 120, 140, 110, 150, 145]
y2 = [90, 80, 120, 150, 170, 140]
```

TITOLO ED ETICHETTE ASSI

```python
plt.plot(x,y1)
plt.title("Nome del grafico") #TITOLO GRAFICO
plt.xlabel("Et√†") #ETICHETTA X
plt.ylabel("Valore") #ETICHETTA Y
plt.show()
```

GRAFICO CON DUE LINEE

```python
#GRAFICO CON DUE LINEE
plt.plot(x,y1)
plt.plot(x,y2)
plt.title("Nome del grafico") 
plt.xlabel("Et√†") 
plt.ylabel("Valore") 
plt.legend(["italiani", "francesi"]) #LEGENDA
#Python capisce da solo che, siccome ho due plot(y1 e y2), la prima etichetta della legenda va a y1
#Questo ci complica un po' le cose se volessimo fare pi√π plot 
#perch√® dovremmo ricordarci a quali valori corrispondono le etichette della legenda
#prossimo riquadro mostra una via semplificata

plt.show()
```

LEGENDA SEMPLIFICATA

```python
#GRAFICO CON DUE LINEE E LEGENDA SEMPLIFICATA
plt.plot(x,y1, label = "Italiani")
plt.plot(x,y2, label = "Francesi")
#basta usare label = "nome dato"

plt.title("Nome del grafico") 
plt.xlabel("Et√†") 
plt.ylabel("Valore") 
plt.legend() #LEGENDA

plt.show()
```

GRIGLIA

```python
#AGGIUNGIAMO UNA GRIGLIA
plt.plot(x,y1, label = "Italiani")
plt.plot(x,y2, label = "Francesi")
plt.title("Nome del grafico") 
plt.xlabel("Et√†") 
plt.ylabel("Valore") 
plt.legend() 
plt.grid() #GRIGLIA


plt.show()
```

MODIFICARE COLORE LINEE, LO STILE E LA DIMENSIONE

```python
#MODIFICHIAMO IL COLORE DELLE LINEE
plt.plot(x,y1, label = "Italiani", color = "red", linewidth = 5)
plt.plot(x,y2, label = "Francesi", color = "#212121", linestyle = "dotted")
#basta usare color = "nome colore" oppure color = "#numero colore"
#linewidth serve a settare lo spessore
#linestyle per settare la stile della linea(si potrebbe scrivere anche linestyle = "---")

plt.title("Nome del grafico") 
plt.xlabel("Et√†") 
plt.ylabel("Valore") 
plt.legend() 
plt.grid() 


plt.show()
```

AGGIUNGIAMO DEI MARCATORI

```python
plt.plot(x,y1, label = "Italiani", color = "red", marker = "o")
#comando marker = "tipo di marcatore"
plt.plot(x,y2, label = "Francesi", color = "#212121", linestyle = "dotted")

plt.title("Nome del grafico") 
plt.xlabel("Et√†") 
plt.ylabel("Valore") 
plt.legend() 
plt.grid() 


plt.show()
```

STILI GRAFICO PREDEFINITI
