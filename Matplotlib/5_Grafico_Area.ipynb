GRAFICO A LINEE RIEMPITO:
- plot a linee, riempire area, alpha
- riempire tra linea e punto, aggiungere condizione

```python
from matplotlib import pyplot as plt
plt.style.use("ggplot")
```

```python
x = [25, 26, 27, 28 , 29, 30]
y = [80, 150, 170, 130, 140, 190]

plt.plot(x,y)

plt.title("Nome grafico")
plt.show

#l'obiettivo Ã¨ colorare l'area sotto la linea
```

```python
x = [25, 26, 27, 28 , 29, 30]
y = [80, 150, 170, 130, 140, 190]

plt.plot(x,y)
#usiamo il comando fill_between(). Tra parentesi inseriamo tra dove e dove fare il riempimento
#Possiamo anche definire un colore con color=, l'opacitÃ  con alpha=(float), ecc
plt.fill_between(x,y, color = "blue", alpha = 0.5)

plt.title("Nome grafico")
plt.show
```

RIEMPIRE DA UN CERTO PUNTO AD UN ALTRO

```python
x = [25, 26, 27, 28 , 29, 30]
y = [80, 150, 170, 130, 140, 190]
media = 140
#valore di riferimento

plt.plot(x,y)
plt.fill_between(x,y, media, color = "blue", alpha = 0.5)

plt.title("Nome grafico")
plt.show
```

AGGIUNGERE CONDIZIONE

```python
import numpy as np
#dobbiamo importare numpy perchÃ¨ non possiamo fare la condizione direttamente sulla lista
#ma ci serve un array

x = np.array([25, 26, 27, 28 , 29, 30])
y = np.array([80, 150, 170, 130, 140, 190])
media = 140

plt.plot(x,y)
#la condizione viene definita con where = (condizione). Interpolate=True ci permette di colorare l'area
plt.fill_between(x,y, media, where=(y>150), color = "blue", alpha = 0.5, interpolate = True)

plt.title("Nome grafico")
plt.show
```

```python
#altro esempio di condizione senza medi

x = np.array([25, 26, 27, 28 , 29, 30])
y = np.array([80, 150, 170, 130, 140, 190])

plt.plot(x,y)
plt.fill_between(x,y, where=(y>150), color = "blue", alpha = 0.5, interpolate = True)

plt.title("Nome grafico")
plt.show
```
