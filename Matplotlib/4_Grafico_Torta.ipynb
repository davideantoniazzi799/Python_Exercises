GRAFICO A TORTA:
- crearlo, aggiungere etichette, bordi
- colori, explode, startangle, percentuale

```python
from matplotlib import pyplot as plt
```

CREARLO

```python
#il comando Ã¨ plt.pie()
slices = [50, 50]
plt.pie(slices)

#la somma totale non deve essere per forza 100. Se avessimo slices =[45, 89] matplot calcolerebbe da solo
#la proporzione e farebbe il grafico su di esse

plt.title("Nome grafico")
plt.show
```

AGGIUNGIAMO LE ETICHETTE

```python
#per crearle noi, dobbiamo prima creare l'oggetto labels con le etichette con lunghezza uguale a quella dei dati
#poi in plt.pie dobbiamo fare labels=labels(o nome variabile delle etichette)
slices = [60, 30]
labels = ["Si", "No"]
plt.pie(slices, labels = labels)


plt.title("Nome grafico")
plt.show
```

CAMBIAMO I COLORI

```python
#funziona come con labels, solo con colors = nome variabile colori
slices = [60, 30]
labels = ["Si", "No"]
colors = ["red", "yellow"]
plt.pie(slices, labels = labels, colors=colors)

plt.title("Nome grafico")
plt.show
```

BORDI

```python
slices = [60, 30]
labels = ["Si", "No"]

plt.pie(slices, labels = labels, wedgeprops={"edgecolor": "#000000"})
#wedgeprops Ã¨ un dizionario che ci permette di modificare molte caratteristiche dei bordi
#con edgecolor cambiamo il colore ad esempio

plt.title("Nome grafico")
plt.show
```

CAMBIARE L'ANGOLO

```python
slices = [60, 30]
labels = ["Si", "No"]

plt.pie(slices, labels = labels, wedgeprops={"edgecolor": "#000000"}, startangle=90)
#startangle ci permette di modificare l'angolo da cui parte il grafico

plt.title("Nome grafico")
plt.show
```

DARE RISALTO A DELLE FETTE: EXPLODE

```python
#usiamo EXPLODE che funzione come le labels o colori, con numeri float
slices = [60, 30]
labels = ["Si", "No"]
explode = [0, 0.1]

plt.pie(slices, labels = labels, wedgeprops={"edgecolor": "#000000"}, startangle=90, explode = explode)

plt.title("Nome grafico")
plt.show
```

PERCENTUALI

```python
slices = [60, 30]
labels = ["Si", "No"]

#dobbiamo usare autopct="%1.1f%%" e ci mostrare come il grafico Ã¨ diviso in percentuale
#%1.1f%% Ã¨ una format string
plt.pie(slices, labels = labels, wedgeprops={"edgecolor": "#000000"}, 
        startangle=90, autopct = "%1.1f%%")

plt.title("Nome grafico")
plt.show
```
