GRAFICO DISPERSIONE:
- grafico base, dimensione punti s, colore c, marker, edgecolor, linewidth, alpha
- intensità colori come dati, cmap e colorbar, dimensione come dati 

```python
from matplotlib import pyplot as plt
plt.style.use("ggplot")
```

GRAFICO BASE

```python
#il comando è plt.scatter(x,y)
x = [1,5,6,8,3,2,2,7,9,1]
y = [8,8,4,6,2,7,1,3,1,7]

plt.scatter(x, y)

plt.title("Nome grafico")
plt.show
```

PERSONALIZZAZIONE

```python
x = [1,5,6,8,3,2,2,7,9,1]
y = [8,8,4,6,2,7,1,3,1,7]

#con s= possiamo modificare la dimensione dei punti

#c ci permette di cambiare il colore dei PUNTI

#marker = "x" ci permette di cambiare come è fatto il punto. Si possono anche usare interi
#ma sono associati a forme già stabilite

#edgecolor ci permette di modificare il colore dell'intorno del punto

#linewidth ci permette di aumentare lo spessore del bordo del punto

#alpha ci permette di definire l'opacità. Valore in decimali

plt.scatter(x, y, s= 100, c="green", edgecolor="yellow", linewidth = 2)

plt.title("Nome grafico")
plt.show
```

INTENSITA' COLORE CON DATI

```python
#possiamo definire l'intesità del colore in base ai dati
x = [1,5,6,8,3,2,2,7,9,1]
y = [8,8,4,6,2,7,1,3,1,7]
colors = [2,4,6,6,1,9,4,6,2,8]
#in questo esempio abbiamo associato più colori ai nostri dati

plt.scatter(x, y, s= 100, c=colors)

plt.title("Nome grafico")
plt.show
```

```python
#per la vera intesità usiamo cmap
x = [1,5,6,8,3,2,2,7,9,1]
y = [8,8,4,6,2,7,1,3,1,7]
colors = [2,4,6,6,1,9,4,6,2,8]
#otteniamo una diversa intensità di verdi

plt.scatter(x, y, s= 100, c=colors, cmap="Greens")
cbar = plt.colorbar()
cbar.set_label("Intensità")
#questo comando di permette di visualizzare la barra che ci mostra l'intensità
#serve creare una barra con plt.colorbar() e nominarla con .set_label("nome")
#in questo caso più è scuro il verde, maggiore è l'intensità

plt.title("Nome grafico")
plt.show
```

DIMENSIONE PUNTO

```python
from matplotlib import pyplot as plt
plt.style.use("classic")

x = [1,5,6,8,3,2,2,7,9,1]
y = [8,8,4,6,2,7,1,3,1,7]
colors = [2,4,6,6,1,9,4,6,2,8]
sizes = [200,250,120,500, 430, 700, 567, 312, 691, 443]
#possiamo vedere come la dimensione dei punti varia

plt.scatter(x, y, s= sizes, c=colors, cmap="Greens")
cbar = plt.colorbar()
cbar.set_label("Intensità")


plt.title("Nome grafico")
plt.show
```
