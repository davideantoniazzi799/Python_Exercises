ISTOGRAMMA:
- basicon con bins, bins personalizzati, color e edgecolor
- axvline

```python
from matplotlib import pyplot as plt
plt.style.use("ggplot")
```

BINS: sono le "scatole" dove aggregriamo i nostri valori

```python
#il comando Ã¨ plt.hist con uso di bins per decidere quante barre
x = [13, 18, 18, 22, 25, 27, 29, 29, 30, 35, 36, 37, 40, 41, 45, 45,54, 56, 65]
plt.hist(x,bins = 6)
#le barra rappresentano quanti valori sono compresi tra 21-30, 31-40, ecc

plt.title("Nome grafico")
plt.show
```

BINS PERSONALIZZATI

```python
x = [13, 18, 18, 22, 25, 27, 29, 29, 30, 35, 36, 37, 40, 41, 45, 45,54, 56, 65]
bins = [10, 20, 30, 40, 50, 60, 70] #personalizzati
plt.hist(x,bins = bins, color="grey", edgecolor = "black") #colore barre e linee personalizzato

plt.title("Nome grafico")
plt.show
```

AXVLINE: linea verticale. Ax = axis, vline = vertical line

```python
x = [13, 18, 18, 22, 25, 27, 29, 29, 30, 35, 36, 37, 40, 41, 45, 45,54, 56, 65]
bins = [10, 20, 30, 40, 50, 60, 70]
plt.hist(x,bins = bins, color="grey", edgecolor = "black") 
plt.axvline(27, color="red", linewidth="3")
#ci permette di visualizzare una linea verticale sul valore 27 con colore e spessore personalizzati

plt.title("Nome grafico")
plt.show
```
