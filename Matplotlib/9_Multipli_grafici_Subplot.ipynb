SUBPLOT:
- utilizzare subplots con 1 grafico, 2 e più grafici, 2 e più finestre

```python
from matplotlib import pyplot as plt
plt.style.use("ggplot")
```

```python
x = [1,2,3,4,5]
y = [3,6,7,2,5]

plt.plot(x,y)

plt.title("Nome grafico")
plt.xlabel("Prova")
plt.show
```

PIU' GRAFICI

```python
#dobbiamo introdurre il concetto di fig(finestra) e ax(assi)
#con plt.subplots() possiamo creare un'altra finestra usando fig basata su ax
#ora invece di plt possiamo ax quando richiamiamo altre funzioni
#queste funzione devono essere precedute da set_nomefunzione

fig, ax = plt.subplots()
x = [1,2,3,4,5]
y = [3,6,7,2,5]

ax.plot(x,y)

ax.set_title("Nome grafico")
ax.set_xlabel("Prova")
plt.show

#in pratica abbiamo creato un oggetto a parte 
```

PIU' GRAFICI NELLA STESSA FIGURA

```python
fig, (ax1, ax2) = plt.subplots(nrows = 2, ncols = 1)
#questo crea due grafici un due righe e una colonna
#l'uso di ax1 e ax2 è essenziale per poter accedere ad uno dei due grafici

#per modificare uno dei grafici, ci basterà usare ax1 o ax2
#quando richiamiamo la funzione
x = [1,2,3,4,5]
y = [3,6,7,2,5]

ax1.plot(x,y)
ax1.set_title("Nome grafico")
ax1.set_xlabel("Prova")

ax2.plot(x,y)
ax2.set_title("Nome grafico")
ax2.set_xlabel("Prova")

plt.show
```

```python
fig, (ax1, ax2) = plt.subplots(nrows = 2, ncols = 1, sharex=True)
#sharex ci permette di condividere l'asse delle x

x = [1,2,3,4,5]
y = [3,6,7,2,5]

ax1.plot(x,y)
ax1.set_title("Nome grafico")

ax2.plot(x,y)
ax2.set_xlabel("Prova")

plt.show
```

CREARE PIU' FINESTRE

```python
#basta duplicare
fig, ax1 = plt.subplots()
fig, ax2 = plt.subplots()
#possiamo usare nrows e ncols come vogliamo

x = [1,2,3,4,5]
y = [3,6,7,2,5]

ax1.plot(x,y)
ax1.set_title("Nome grafico")
ax1.set_xlabel("Prova")

ax2.plot(x,y)
ax2.set_title("Nome grafico")
ax2.set_xlabel("Prova")

plt.show
```
