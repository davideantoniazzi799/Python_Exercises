GRAFICO A BARRE:
- crearlo, aggiungere secondo grafico
- barre affiancate, grafico orizzontale

```python
from matplotlib import pyplot as plt
```

CREARLO

```python
x = [25, 26, 27, 28, 29, 30]
y1 = [100, 120, 140, 110, 150, 145]
y2 = [90, 80, 120, 150, 170, 140]
```

```python
#invece di .plot() usiam .bar()

plt.bar(x,y1, label = "Italiani", color = "red")
#plt.plot(x,y2, label = "Francesi", color = "#212121")

plt.title("Nome del grafico") 
plt.xlabel("Età") 
plt.ylabel("Valore") 
plt.legend() 
plt.grid() 


plt.show()
```

BARRE AFFIANCATE

```python
#potremmo farne anche due ma in questo modo si sovrappongono. Il prossimo codice ha un'opzione
plt.bar(x,y1, label = "Italiani", color = "red")
plt.bar(x,y2, label = "Francesi", color = "#212121")

plt.title("Nome del grafico") 
plt.xlabel("Età") 
plt.ylabel("Valore") 
plt.legend() 

plt.show()
```

```python
import numpy as np
#la soluzione comporta l'uso di numpy e l'indicizzazione dei valori della x.
#gli indici associati ai valori sull'asse x verranno usati per creare il grafico
#noi comunque vogliamo visualizzare i valori di x sul grafico

#introduciamo quindi il concetto di LARGHEZZA
#la larghezza ci servira sia per definire la width delle barre
#ma anche per riuscire a separare i due grafici e metterli vicini
#lo facciamo con indexes+width
#a questo punto dobbiamo centrare i ticks dell'asse x e inserire i veri valori di x, non i suoi indici

indexes = np.arange(6)
width = 0.3
plt.bar(indexes,y1, label = "Italiani", color = "red", width=width)
plt.bar(indexes+width,y2, label = "Francesi", color = "#212121", width=width)

plt.title("Nome del grafico") 
plt.xlabel("Età") 
plt.ylabel("Valore") 
plt.legend() 

#questo ci permette di posizionare il valore al centro delle due barre. 
#Tra parentesi è calcolata la posizione e con x andiamo a nominare i tick della x con i suoi valori
plt.xticks(indexes+width/2, x)

plt.show()
```

BARRE ORIZZONTALI

```python
#il comando è plt.barth()
plt.barh(x,y1, label = "Italiani", color = "red")

plt.title("Nome del grafico") 
plt.xlabel("Età") 
plt.ylabel("Valore") 
plt.legend() 

plt.show()
```
