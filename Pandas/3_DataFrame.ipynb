DATAFRAME: come si crea e come leggere con loc

```python
import pandas as pd

ds = {
    "Mesi": ["Gennaio", "Febbraio", "Marzo"],
    "Giorni": [31, 28, 31],
    "Stagione": ["Inverno", "Inverno", "Primavera"],
    "Festivita": ["Capodanno", "Carnevale", "Festa delle Donne"]
}
#creare DataFrame
df = pd.DataFrame(ds)
df
```

SELEZIONARE UNA RIGA CON LOC (LOCALIZATION)

```python
#Il comando print(df[0]) ci da errore
#Dobbiamo usare .loc()
print(df.loc[0])
```

```python
#Possiamo anche prendere pi√π liste (occhio alle quadrate):
print(df.loc[[0,1]])
```

CAMBIARE L'INDICE DI UNA RIGA

```python
df = pd.DataFrame(ds, index = ["Riga1", "Riga2", "Riga3"])
df
```

```python
#Se volessimo usare loc ora:
print(df.loc[["Riga3"]])
```
