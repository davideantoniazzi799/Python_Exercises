AGGIUNGERE, RIMUOVERE E SPOSTARE COLONNE
- aggiungere: una o più con nome, insert(indice, nome, valore), loc[:,"Nome"] = valore
- assign(nome = valore)
- rimuovere con drop, del e pop
- muovere on print loc, iloc, reverse, subset di colonne, df.columns.tolist()

```python
import pandas as pd
df = pd.read_csv("pokemon.csv")
```

Tutti gli esempi sono fatti con print per NON cambiare il dataframe in modo permanente.
Se volessimo cambiarlo permanentemente, basta assegnarlo ad una nuova variabile o a df di nuovo.

AGGIUNGERE

```python
#una colonna
#la sintassi è:
#dataframe["Nome_nuova_colonna"] = valore
#Questo aggiunge la "nuova colonna" in coda con il valore assegnato al dataframe

df["Wiii"] = "ciao"
df
```

```python
#più colonne
#la sintassi è:
#dataframe[["Nome_nuova_colonna", "Nome_nuova_colonna", ecc]] = [valore, valore, ecc]
#Questo aggiunge le "nuova colonne" in coda con il valore assegnato al dataframe

df[["Wi","Switch","Gameboy"]] = ["2nd","7th","1st"]
df
```

```python
#usiamo INSERT
#la sintassi è:
#dataframe.insert(indice,"Nome",valore)
#Questo inserisce la "nuova colonna" nell'indice, con il nome e con il valore scelti

df.insert(1,"Nome","abcdef") #l'indice 0 è dell'ID
df
```

```python
#usiamo LOC
#la sintassi è:
#dataframe.loc[: ,"Nome_colonna_nuova"] = valore
#Questo inserisce la "nuova colonna" in coda con il nome e con il valore scelti.
#I : servono a dire che la colonna va aggiunta per tutte le righe

df.loc[:, "Wii"] = "abcdef"
df
```

```python
#usiamo ASSIGN
#la sintassi è:
#dataframe_modificato = dataframe.assign(Nome_colonna = valore)
#Questo inserisce la "nuova colonna" in coda con il nome e con il valore scelti.
#Dobbiamo usare = per assegnare questo cambiamento ad un nuovo dataframe e il nome colonna non necessita ""

df = df.assign(Wii = "abcdef")
df
```

RIMUOVERE

```python
#usiamo DROP
#la sintassi è:
#dataframe.drop("Nome_colonna_rimuovere", axis = 1)
#Questo rimuove la colonna selezionata. Axis è necessario per selezionare le colonne
#Con inplace=true la rimuove direttamente dal dataframe che abbiamo, altrimenti ne crea un altro.

df.drop("Legendary", inplace=True, axis = 1)
df
```

```python
#usiamo DEL
#la sintassi è:
#del dataframe["Nome_colonna_rimuovere"]
#Questo rimuove la colonna selezionata.

del df["Legendary"]
df
```

```python
#usiamo POP
#la sintassi è:
#colonna_rimossa = dataframe.pop("Nome_colonna_rimuovere")
#Questo rimuove la colonna selezionata e la prende in una variabile esterna

colPop = df.pop("Legendary")
df.head(2)
colPop.head(2)
```

MUOVERE UNA COLONNA

```python
#con LOC quando stampiamo
#la sintassi è:
#print(dataframe.loc[:,["Nomi_colonne"]])
#Questo ci permette di prendere le colonne in ordine sparso e invertirle
#Con : prendiamo tutte le righe

print(df.loc[:,["Type 1", "Name", "Total", "Attack"]])
```

```python
#con ILOC quando stampiamo
#la sintassi è:
#print(dataframe.iloc[:,[indici_colonne]])
#Questo ci permette di prendere le colonne in ordine sparso e invertirle usando l'indice delle colonne
#Con : prendiamo tutte le righe

print(df.iloc[:,[1, 2, 5, 3]]) 
#il problema di iloc è ricordarsi a quale colonna corrisponde l'indice
```

```python
#SUBSET di colonne
#la sintassi è:
#print(dataframe[["Nomi colonne sparse"]])
#Questo ci permette di prendere le colonne in ordine sparso

print(df[["Type 1", "Attack", "Name"]]) 
```

```python
#usiamo REVERSE: inverte l'indice delle colonne
#sintassi:
#variabile = list(dataframe.columns)
#variabile.reverse()
#dataframe[variabile]

columns = list(df.columns)
columns.reverse()
df[columns]
```

```python
#usiamo df.columns.tolist(): inverte l'indice delle colonne come sopra, solo in modo semplificato
#sintassi:
#variabile = dataframe.columns.tolist()
#variabile.reverse()
#dataframe[variabile]

columns = df.columns.tolist()
columns.reverse()
df[columns]
```
