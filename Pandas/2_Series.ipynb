SERIES:
Creare una series e prendere dati per indice, aggiungere e prendere label, 
creare una serie da un dizionario e prendere solo alcuni item

```python
import pandas as pd

ds = {
    "Mesi": ["Gennaio", "Febbraio", "Marzo", "Aprile", "Maggio"],
    "Giorni": [31, 28, 31, 30, 31]
}

df = pd.DataFrame(ds)
```

La Series corrisponde ad una colonna del dataframe. In particolare è un array colonna.
Come creare una Series?

```python
#ci serve la funzione pandas.Series()
ds = [1,2,3,4,5]
print(ds)
series = pd.Series(ds)
print(series)
#Potete vedere la differenza tra una liste e una pandas Series
```

Selezionare i dati di una Series: indici

```python
#Indice
print(series[0])  # Primo elemento, si parte da 0
```

Cambiare la label degli indici

```python
#Usiamo l'opzione index e creiamo una serie con etichette personalizzate lunga quanto la lista ds
series = pd.Series(ds, index=["a", "b", "c", "d", "e"])
print(series)
```

Creare una serie direttamente da un dict

```python
#Usiamo pd.Series() per creare una serie da un dizionario
ds = {"a": 1,"b":2, "c" : 3, "d": 4, "e" : 5}
serie = pd.Series(ds)
print(serie)
```

```python
#Partendo da un dizionario, possiamo prendere solo alcuni item
#Questo perchè i dizionari hanno le chiavi
ds = {"a": 1,"b":2, "c" : 3, "d": 4, "e" : 5}
serie = pd.Series(ds, index = ["a", "b"])
print(serie)
```
