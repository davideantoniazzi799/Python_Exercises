FILTRARE DATI:
- uguale, diverso, in, str.contains, ><, &, |, ~
- query, loc

```python
import pandas as pd

df = pd.read_csv("pokemon.csv")
```

```python
#UGUALE
#quella interna è la condizione. Questa sintassi è ottimale quando abbiamo una sola colonna
print(df[df.Name == "Bulbasaur"])
```

```python
#DIVERSO
(df[df.Name != "Bulbasaur"])
```

```python
#STR.CONTAINS
#ci filtra nella colonna per quei valori che hanno una parte di stringa che vogliamo noi
#per questo dobbiamo usare dataframe["nomecolonna"]

df[df["Name"].str.contains("saur")]
```

```python
#CON FILTRO + ISIN
#ci da solo le righe che hanno quel nome
filtro = ["Bulbasaur", "Squirtle"]
df[df["Name"].isin(filtro)]
```

```python
#MAGGIORE (MINORE ANALOGO)
df[df["Total"] > 700]
```

```python
#Usiamo la ~ : ci da il contrario di quello che richiede la condizione
df[~df["Total"] < 200] #restituice tutti quelli sopra a 200 di Total
```

```python
#Usiamo AND &
df[(df["Total"] > 700) & (df["Sp. Atk"] > 160)] #occhio alle parentesi!
```

```python
#Usiamo OR |
df[(df["Total"] > 760) | (df["Total"] < 250)] #occhio alle parentesi!
```

```python
#usiamo LOC
#filtriamo le righe e non le colonne! E occhio alle parentesi!
df.loc[(df["Name"] == "Bulbasaur")] #ci da la riga la cui colonna Name == "Bulbasaur" 
```

```python
df.loc[(df["Name"] == "Bulbasaur"), ["HP", "Name", "Generation"]] 
#ci da la riga la cui colonna Name == "Bulbasaur" con solo le colonne che vogliamo. Vale anche per altri filtri
```

```python
#usiamo QUERY  
# basta mettere la condizione tra tonde e occhio con le stringhe alle virgolette 
# Problema con query: non possiamo usare colonne il cui nome contiene spazi vuoti!
# Dobbiamo eliminarli in qualche modo 
df.query('Name == "Bulbasaur"') 
df.query('Total > 750')
```

```python
df.query('Total > 750 and Speed > 100')
```
