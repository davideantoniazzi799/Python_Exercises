LOC E iLOC:
- loc: riga, cella, più righe e più colonne
- iloc

```python
import pandas as pd
df = pd.read_csv("pokemon.csv", index_col = 1) #così si prende la colonna 1 come colonna che da l'indice alla riga
df
```

LOC: localization cioè andare a prendere degli elementi

Loc usa due elementi: il numero della riga e il nome della colonna che ci interessano

iLOC = index localization.
Usa due elementi entrambi interi: gli indici della riga e della colonna

```python
df.loc["Bulbasaur"] #questo ci da la riga corrispondente a Bulbasaur
#df.iloc["Bulbasaur"] non funzionerebbe perchè non abbiamo fornito l'indice
```

```python
df.iloc[0] #questo invece funziona
```

```python
df.loc["Bulbasaur", "Total"] 
#questo ci da un valore preciso corrispondente alla riga Bulbasaur e alla colonna Total
```

```python
df.iloc[0, 3] 
#stesso risultato ma modo diverso con iloc (il numero della colonna cala di uno per index_col = 1)
```

PRENDERE PIU' RIGHE

```python
df.loc["Bulbasaur":"Venusaur"] #slicing con loc(prende anche l'estremo)
```

```python
df.iloc[0:3] #slicing con iloc(non prende l'estremo superiore!)
```

```python
df.loc[:, ["Type 1", "Type 2"]]
#prende tutte le righe e due colonne con loc
```

```python
df.iloc[:, [1,2]]
#prende tutte le righe e due colonne con iloc
```
