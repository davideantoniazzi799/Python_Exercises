ITERARE DATAFRAME:
- items, iterrows e itertuples

```python
import pandas as pd
df = pd.read_csv("pokemon.csv").head(1)
```

- ITEMS(una volta iteritems) serve per ottenere chiavi:valore, come fosse un dizionario
- ITERROWS funziona per righe tramite l'indice
- ITERTUPLES ci ritorna tutta la riga come se fosse una tupla, prendendo anche il nome della colonna

Tutte le iterazioni creano delle copy dei dataframe, quindi non li modificano

```python
#items:
#la sintassi è:
#for key:value in DataFrame.items():
#   print(key)
#quindi ci restituisce la riga come se fosse un dizionario

for key,value in df.items():
    print(key) #questo stampa le chiavi

#ci ha stampato il nome delle colonne, questo perchè abbiamo Name:Bulbasaur, Type 1: Grass, ecc
#se avessimo avuto head(2) avremmo ottenuto lo stesso risultato
```

```python
for key,value in df.items():
    print(value) #questo ci printa i valori
#ci stampa anche il nome e il tipo della chiave(le colonne) con l'indice dell'elemento
#se avessimo head(2) otteniamo risultati diversi(vedi sotto)
```

```python
df = pd.read_csv("pokemon.csv").head(2)
for key,value in df.items():
    print(value)
```

```python
#iterrows
#la sintassi è:
#for index:row in DataFrame.iterrows():
#   print(index)

for index,row in df.iterrows():
    print(index) #ci restituisce l'indice degli elementi del dataframe
```

```python
for index,row in df.iterrows():
    print(index, row) #ci stampa l'indice della riga e anche ogni riga di ogni elemento del dataframe
```

```python
#itertuples
#la sintassi è:
#for row in DataFrame.iterrows():
#   print(row)
for row in df.itertuples():
    print(row)
#per ogni riga del dataframe, ci restituisce la riga in formato tupla
#notiamo che alcuni nome di colonna sono _numero. Questo perchè con itertuples alcuni caratteri o
#gli spazi non riescono ad essere letti
```
