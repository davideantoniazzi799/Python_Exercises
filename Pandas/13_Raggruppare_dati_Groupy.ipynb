RAGGRUPPARE AGGREGATI CON GROUPY:
- raggruppare per una colonna e colonne multiple
- iterare gruppi
- sum, mean, count, min, max, aggiungere colonna count

```python
import pandas as pd

df = pd.read_csv("pokemon.csv")
```

Raggruppare per una colonna o più colonne

```python
#UNA colonna
types = df.groupby("Type 1")
types.groups
#con groupby andiamo a creare un oggetto(types) il quale ha l'attributo .groups
#con questo attributo possiamo vedere a quale gruppo appartengono i vari elementi, in formato dict
```

```python
#questo è come possiamo fare un ciclo sui gruppi
for name,group in types:
    print(name)
    print(group)
```

```python
#PIU' colonne
types = df.groupby(["Type 1", "Type 2"])
#basta usare una lista delle colonne che ci interessano per fare il raggruppamento

types.groups
```

```python
for name,group in types:
    print(name)
#in pratica sono tutte le combinazioni disponibili
```

STATISTICHE AVANZATE

```python
#MEDIA
types = df.groupby("Type 1").mean(numeric_only=True)
#quando creiamo l'oggetto types ci basta aggiungere .mean alla fine del comando
#numeric_only=True serve per calcore la media solo sulle variabili numeriche
#perchè di default è false e quindi calcolerebbe anche nelle variabili non numeriche dandoci errore

#la media è calcolata per gruppo!

types
```

```python
types = df.groupby("Type 1").mean(numeric_only=True)[["HP", "Attack", "Defense"]]
#con questo comando andiamo a selezionare la media delle sole colonne che ci interessano

types
```

```python
types = df.groupby("Type 1").mean(numeric_only=True)[["HP", "Attack", "Defense"]].sort_values(by= "Defense")
#con questo comando andiamo a selezionare la media delle sole colonne che ci interessano
#e ad ordinare secondo la difesa (più bassa a più alta, altrimetni ascending = False)

types
```

```python
#SOMMA
types = df.groupby("Type 1").sum(numeric_only=True)
#quando creiamo l'oggetto types ci basta aggiungere .sum alla fine del comando
#numeric_only=True serve per calcore la somma solo sulle variabili numeriche
#perchè di default è false e quindi calcolerebbe anche nelle variabili non numeriche dandoci un risultato strano

#la somma è calcolata per gruppo!

types
```

```python
#MINIMO
types = df.groupby("Type 1").min(numeric_only=True)
#quando creiamo l'oggetto types ci basta aggiungere .min alla fine del comando
#numeric_only=True serve per calcore il minimo solo sulle variabili numeriche
#perchè di default è false e quindi calcolerebbe anche nelle variabili non numeriche dandoci errore

#il minimo è calcolato per gruppo!

types
```

```python
#MASSIMO
types = df.groupby("Type 1").max(numeric_only=True)
#quando creiamo l'oggetto types ci basta aggiungere .max alla fine del comando
#numeric_only=True serve per calcore il massimo solo sulle variabili numeriche
#perchè di default è false e quindi calcolerebbe anche nelle variabili non numeriche dandoci errore

#il massimo è calcolato per gruppo!

types
```

```python
#COUNT
types = df.groupby("Type 1").count()
#quando creiamo l'oggetto types ci basta aggiungere .count alla fine del comando

#il count è calcolato per gruppo!

types
```

```python
#COLONNA COUNT
df["count"] = 1
types = df.groupby("Type 1").count()["count"]

#creando una colonna count e richiamandola dopo l'aggregazione dei gruppi abbiamo il conto totale

types
```

```python
#COLONNA COUNT
df["count"] = 1
types = df.groupby(["Type 1", "Type 2"]).count()["count"]

#creando una colonna count e richiamandola dopo l'aggregazione dei gruppi abbiamo il conto totale

types
```
