SALVARE FILE IN:
- csv, csv compresso
- excel, scegliere nome foglio, fogli multipli, append to file

```python
import pandas as pd
from typing import Dict
df = pd.read_csv("pokemon.csv")
df2 = df[["Name", "ID", "Total"]]
df3 = df[["Name", "HP", "Attack", "Defense"]]
```

CSV

```python
#facciamo CSV
#la sintassi è:
#dataframe.to_csv("nome_salvataggio.csv")

df2.to_csv("mini_pokemon.csv") #questo file contiene una colonna indice extra però
#df2.to_csv("mini_pokemon.csv", index= False) questo non avrà quella colonna indice
```

```python
#facciamo CSV compresso
#la sintassi è:
#variabile_compressa = dict(method = "zip", archive_name = "nuovi_pokemon.csv")
#dataframe.to_csv("nome_salvataggio.zip", compression = variabile_compressa)

#ci serve la classe Dict dalla libreria typing

compressed_options = dict(method = "zip", archive_name = "nuovi_pokemon.csv")
df2.to_csv("nuovi_pokemon.zip", index = False, compression=compressed_options)
```

EXCEL

```python
#facciamo EXCEL
#la sintassi è:
#dataframe.to_excel("nome_salvataggio.xlsx")

df2.to_excel("mini_pokemon.xlsx") #questo file contiene una colonna indice extra però
#df2.to_excel("mini_pokemon.xlsx", index= False) questo non avrà quella colonna indice
```

```python
#facciamo EXCEL con nome foglio
#la sintassi è:
#dataframe.to_excel("nome_salvataggio.xlsx", sheet_name = "Nome")

df2.to_excel("mini_pokemon.xlsx", index= False, sheet_name="Pokemon_Totali")
```

```python
#facciamo EXCEL con più fogli
#la sintassi è:
#with pd.ExcelWriter("nome_file_salvataggio.xlsx") as writer:
# dataframe1.to_excel(writer, sheet_name="nome foglio 1")
# dataframe2.to_excel(writer, sheet_name="nome foglio 2") e così via

#come se fosse un ciclo for

with pd.ExcelWriter("nuovi_pokemon.xlsx") as writer:
    df2.to_excel(writer, sheet_name="foglio_1", index=False)
    df3.to_excel(writer, sheet_name="foglio_2", index=False)
```

APPENDERE AD UN FILE GIA' ESISTENTE

```python
df2 = df.iloc[0:10, 0:4]
```

```python
#la sintassi è:
#with pd.ExcelWriter("nome_file_salvataggio.xlsx", mode = "a") as writer:
# dataframe1.to_excel(writer, sheet_name="nome foglio 1")
# dataframe2.to_excel(writer, sheet_name="nome foglio 2") e così via

#la mode = a significa append, molto importante

with pd.ExcelWriter("nuovi_pokemon.xlsx", mode = "a") as writer:
    df2.to_excel(writer, sheet_name="foglio_appeso", index=False)
```
