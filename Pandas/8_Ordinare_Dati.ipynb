ORDINARE I DATI:
- per indice crescente e decrescente, per colonna e per colonne

```python
import pandas as pd
df = pd.read_csv("pokemon.csv")
```

Quando ordiniamo i dati, ci ritorna un altro dataframe. Quindi dobbiamo raccoglierlo

```python
#ordiniamo per indice crescente

#la sintassi è:
#dataFrame_ordinato = DataFrame.sort_index()

sorted_df = df.sort_index()
sorted_df.head(5)
#ci restituisce il dataframe originale in poche parole
```

```python
#ordiniamo per indice decrescente

#la sintassi è:
#dataFrame_ordinato = DataFrame.sort_index(ascending = False)

sorted_df = df.sort_index(ascending = False)
sorted_df.head(5)
#ci restituisce il dataframe originale con indici in ordine descrescente
```

```python
#ordiniamo per colonna

#la sintassi è:
#dataFrame_ordinato = DataFrame.sort_values(by = "Nome colonna")
#il by ci serve per decidere secondo quale colonna ordinare alfabeticamente
#possiamo anche usare ascending=False così l'ordine è invertito

sorted_df = df.sort_values(by="Name")
sorted_df.head(5)
#ci restituisce il dataframe originale in ordine alfabetico secondo la colonna Name
```

```python
#ordiniamo per colonne

#la sintassi è:
#dataFrame_ordinato = DataFrame.sort_values(by = ["Nome colonna","Nome colonna, ecc"])
#il by ci serve per decidere secondo quali colonne ordinare, le quali vanno inserite come LISTA
#possiamo anche usare ascending=False così l'ordine è invertito per tutte le colonne inserite,
#possiamo anche decidere di usare ascending per una sola colonna o alcune mentre altre non lo usano con
#ascending = [True, False, ecc] con lunghezza pari al numero di colonne che abbiamo nel by

sorted_df = df.sort_values(by=["Total","HP"])
sorted_df.head(5)
#ci restituisce il dataframe originale in ordine e come lo fa?
#ordina la colonna "Total" e quando ci sono dei doppioni ordina per "HP"
```
