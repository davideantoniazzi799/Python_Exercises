SEARCHING, FILTERING AND SORTING

```python
import numpy as np
```

```python
arr = np.array([1,2,3,4,5,6,7])
```

Cercare l'indice di un certo elemento: WHERE

Viene chiamato con np.where() dove inseriamo l'array e cosa vogliamo cercare al suo interno.
Ci ritorno l'indice il cui elemento corrisponde a quello che abbiamo cercato.

```python
arrIndici = np.where(arr == 4)
print(arrIndici)
```

```python
#questo ci ritorna gli indici di tutti gli elementi pari
arrIndici = np.where(arr %2 == 0)
print(arrIndici)
```

SORT

```python
arr = np.array([3,1,5,7,4,2,6])
print(np.sort(arr))  #ritorna una copia ordinata dell'array senza modificare l'originale. 
#Funziona anche con le stringhe e ci ritorna sempre un array ordinato.
```

```python
#Array multidimensionali
arr2D = np.array([[3,1,5],
                  [7,4,2],
                  [6,8,9]])
print(np.sort(arr2D)) #ordina ogni riga
```

FILTRARE: ESEMPIO STATICO, DINAMICO E SCORCIATOIA

```python
arr = np.array([1,2,3,4])
#ESEMPIO STATICO: numeri dispari
filtro = [True, False, True, False]

arrayFiltrato = arr[filtro] #prende solo i valori che sono True
print(arrayFiltrato) 
```

```python
#ESEMPIO DINAMICO: numeri pari
arr = np.array([1,2,3,4])
filtro = []
for numero in arr:
    if numero %2 == 0:
        filtro.append(True) #come con le liste
    else:
        filtro.append(False)

arrayFiltrato = arr[filtro] #prende solo i valori che sono True
print(arrayFiltrato) 
```

```python
#SCORCIATOIA: numeri pari
arr = np.array([1,2,3,4])
filtro = arr%2 == 0
arrayFiltrato = arr[filtro] #prende solo i valori che sono True
print(arrayFiltrato) 
```
