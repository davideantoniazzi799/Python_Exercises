DATA TYPE IN NUMPY
#guardare pagina di w3 school per la documentazione

```python
import numpy as np
arr = np.arange(1, 6)
```

```python
#Per capire il tipo di dato:
print(arr.dtype) # Output: int64 (o int32 a seconda del sistema)
#Quindi .dtype è un attributo che ci dice il tipo di dato degli elementi nell'array NumPy
```

CREARE UN ARRAY DI UN DETERMINATO TIPO

```python
arr = np.array(["1", "2", "3", "4", "5"], dtype = 'S')
print(arr.dtype) # Output: |S1 (stringa di lunghezza 1)
#Quindi abbiamo creato un array di stringhe. Per altri tipi di dato, guardare la documentazione
#https://www.w3schools.com/python/numpy/numpy_data_types.asp
```

CONVERTIRE IL TIPO DI UN ARRAY GIA' CREATO

```python
arr = np.array(["1", "2", "3", "4", "5"])
```

```python
arrInt =arr.astype('i')
# Ora arrInt è un array di interi
print(arrInt.dtype) # Output: int32 o int64 a seconda del sistema
```
