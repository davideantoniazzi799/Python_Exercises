VIEW E COPY

```python
import numpy as np
```

View e copy a livello teorico servono entrambi per copiare il contenuto di un array in ul altro array.
La differenza è:
- copy fa una copia effettiva dell'array e modificando l'array copiato non si va a modificare quello originale.     Quindi siamo proprietari dell'array modificato
- view ci permette solo di vedere l'originale, senza esserne proprietari


```python
# Come creare una COPIA
arr = np.array([1,2,3])
arrCopy = arr.copy()  # copia effettiva

#Quindi con .copy() siamo proprietari dell'array copiato e possiamo modificarlo senza modificare l'originale
```

```python
#Andiamo a modificare l'originale? l'array copiato non cambia
arr[0] = 10
```

```python
print(arr)
print(arrCopy)
```

```python
#come creare una VIEW
arr = np.array([1,2,3])
arrView = arr.view()

#Con .view() non siamo proprietari dell'array e quindi se modifichiamo l'originale, anche la view cambia
```

```python
arr[0] = 10
#Andiamo a modificare l'originale? La View cambia
```

```python
print(arr)
print(arrView)
```

```python
#Per capire se siamo proprietari di un array, possiamo usare l'attributo base, 
# che restituisce None se l'array è proprietario. Se non è proprietario, restituisce l'array originale.
print(arr.base)      # None, perchè è proprietario
print(arrView.base)  # restituisce arr, perchè non è proprietario
```
