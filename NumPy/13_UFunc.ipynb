CREARE LA UFUNC

```python
def add5(x):
    return x + 5

add5 = np.frompyfunc(add5, 1, 1)
#questa operazione trasforma la funzione in una ufunc, quindi è importante per registrarla
#per farlo uso np.frompyfunc(nome_funzione, numero di input, numero di output)
#questo serve per far si che la funzione possa essere applicata a array numpy
#e non solo a singoli valori
#Se avessimo ua lista come input e non usassimo frompyfunc, otterremmo un errore perchè la funzione non è in grado di gestire liste
#con frompyfunc invece la funzione viene trasformata in una ufunc e può essere applicata a array numpy

arr =np.arange(1,6)
print(add5(arr))
```

ADD FUNCTION

```python
arr1 = np.array([10,20,30,40,50])
arr2 = np.array([5,10,15,20,25])
```

```python
print(np.add(arr1, arr2)) #funziona anche con liste
```

SUBTRACT FUNCTION

```python
print(np.subtract(arr1, arr2))
```

MULTIPLY FUNCTION

```python
print(np.multiply(arr1, arr2))
```

DIVIDE FUNCTION

```python
print(np.divide(arr1, arr2))
```

POWER FUNCTION

```python
print(np.power(arr1, arr2))
```

MOD/REMAINDER FUNCTION (resto della divisione)

```python
print(np.remainder(arr1, arr2))
```

TRUNC: per arrotondare al primo intero (uguale a .fix(), around())

CEIL: arrotondare per eccesso

FLOOR: arrotondare per difetto

```python
arr = np.trunc([1.5, 2.765, 3.432, 4.123, 5.567]) #per arrotondare al primo intero (uguale a .fix(), around())
print(arr)
arr = np.ceil([1.5, 2.765, 3.432, 4.123, 5.567]) #arrotondare per eccesso
print(arr)
arr = np.floor([1.5, 2.765, 3.432, 4.123, 5.567]) #arrotondare per difetto
print(arr)
```

FUNZIONE SUM: ci permette di sommare tutti gli elementi di un array

```python
a = np.array([1, 2, 3, 4]) #una dimensione
print(a.sum())
```

```python
b = np.array([[1, 2, 3], 
              [4, 5, 6]]) #due dimensioni
print(b.sum()) #somma tutti gli elementi
print(b.sum(axis=0)) #somma le colonne
print(b.sum(axis=1)) #somma le righe
```

```python
c = np.random.rand(2, 3, 4) #tre dimensioni
print(c.sum(axis=(0, 1))) #somma per riga e colonna
```

FUNZIONE DOT:  ci permette di effettuare l'operazione di moltiplicazione matriciale riga per colonna, 
                quindi il prodotto matriciale. Una matrice ha 2x1 e l'altra 1x2 otteniamo una matrice 2x2.

```python
a = np.array([[1, 2]])
b = np.array([[3], 
              [4]])
print(a.dot(b)) #la matrice di sinistra richiama dot e come argomento ha la matrice di destra
```

```python
print(b.dot(a)) #matrici invertite(occhio alla regola delle dimensioni)
```
