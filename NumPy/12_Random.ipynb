RANDOM: GENERARE UN ARRAY RANDOM CON RANDINT E RAND, CHOICE PER ESTRARRE UN VALORE E GENERARE ARRAY
        DISTRIBUZIONE DATI CON PROBABILITA', SHUFFLE E PERMATUTION

```python
import numpy as np
from numpy import random, array, size
```

```python
#RANDINT
numero = random.randint(100) #numero casuale tra 0 e 100
print(numero)
```

```python
#ARRAY RANDOM CON RANDINT
arr = random.randint(100, size=(15)) #array monodimensionale di 15 numeri casuali tra 0 e 100
print(arr)

arr2 = random.randint(100, size=(2, 10)) #array bidimensionale di 10 numeri casuali tra 0 e 100
print(arr2)
```

```python
#ARRAY RANDOM CON RAND
arr = random.rand(5) #array monodimensionale di 5 numeri casuali tra 0 e 1
print(arr)

#a differenza di randint, con rand non serve specificare il limite massimo perchè è sempre 1

arr2 = random.rand(5, 2) #array bidimensionale di 5 righe e 2 colonne
print(arr2)
```

CHOICE

```python
arr = np.arange(1, 10)
print(random.choice(arr)) #estrae un numero casuale dall'array
```

```python
#Con choice possiamo creare un array composto da numeri estratti casualmente da un altro array
arrChoice = random.choice(arr, size=5) #array monodimensionale con 5 numeri estratti casualmente da arr
print(arrChoice)
```

PROBABILITA'

```python
arr = np.arange(1, 10)
prob = np.array([0.1, 0.2, 0.2, 0.1, 0.1, 0.1, 0.1, 0.05, 0.05]) #probabilità associate a ciascun numero
arrChoice = random.choice(arr, p=prob, size=5)
print(arrChoice)
#quindi usando delle probabilità possiamo influenzare l'estrazione dei numeri casuali
```

SHUFFLE E PERMUTATION

```python
arr = np.arange(1, 10)
```

```python
#SHUFFLE
random.shuffle(arr) #modifica l'array originale mescolandolo, quindi agisce sull'array originale
print(arr)
```

```python
#PERMUTATION
arrPermut = random.permutation(arr) #dobbiamo assegnare il risultato a una nuova variabile
print(arrPermut) #crea una copia dell'array originale mescolata, quindi l'array originale non viene modificato
```
